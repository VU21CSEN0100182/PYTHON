```python
#weight
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35, 25, 25, 15])
fruits=["Apples", "Bananas", "Cherries", "Dates"]
explode=[0.2, 0, 0, 0]
plt.pie(y,labels=fruits,explode=explode)
plt.show() 
```

```python
#new colours
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,15,15,25])
fruits= ["Apples", "Bananas", "Cherries", "Dates"]
colors=["red","yellow","pink","brown"]
plt.pie(y,labels=fruits,colors=colors)
plt.show() 
```

```python
#shadow
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
fruits=["Apples", "Bananas", "Cherries", "Dates"]
explode=[1.3,0,1,2]
plt.pie(y,labels=fruits,explode=explode,shadow=False)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35, 25, 25, 15])
fruits=["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y,labels=fruits)
plt.legend()
plt.show() 
```
